```python
import pandas as pd
```

```python
dataset = pd.read_csv('./../data/clean_chromophore_data.csv')
```

```python
# count unique chromophores
unique_chromophores = dataset['Chromophore'].nunique()
print(f'Number of unique chromophores: {unique_chromophores}')

# count unique solvents
unique_solvents = dataset['Solvent'].nunique()
print(f'Number of unique solvents: {unique_solvents}')
```

```python
# plot distributions of wavelengths
import matplotlib.pyplot as plt
plt.hist(dataset['Absorption max (nm)'], bins=50, color='blue', alpha=0.7)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Frequency')
plt.title('Distribution of Absorption Max Wavelengths')
plt.show()
```

```python
plt.hist(dataset['Emission max (nm)'], bins=50, color='green', alpha=0.7)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Frequency')
plt.title('Distribution of Emission Max Wavelengths')
plt.show()
```

```python
# get dummy MAE if we predict the mean absorption max for all samples
mean_absorption = dataset['Absorption max (nm)'].mean()
mae = (dataset['Absorption max (nm)'] - mean_absorption).abs().mean()
print(f'Dummy MAE (mean predictor) for Absorption max: {mae:.2f} nm')

# get dummy MAE if we predict the mean emission max for all samples
mean_emission = dataset['Emission max (nm)'].mean()
mae_emission = (dataset['Emission max (nm)'] - mean_emission).abs().mean()
print(f'Dummy MAE (mean predictor) for Emission max: {mae_emission:.2f} nm')
```

```python
# dummy on val set
val_dataset = pd.read_csv('./../data/splits/split_70_15_15/val.csv')
mae = (val_dataset['Absorption max (nm)'] - mean_absorption).abs().mean()
print(f'Dummy MAE (mean predictor) for Absorption max on val set: {mae:.2f} nm')
mae_emission = (val_dataset['Emission max (nm)'] - mean_emission).abs().mean()
print(f'Dummy MAE (mean predictor) for Emission max on val set: {mae_emission:.2f} nm')
```

```python
# dummy on test
test_dataset = pd.read_csv('./../data/splits/split_70_15_15/test.csv')
mae = (test_dataset['Absorption max (nm)'] - mean_absorption).abs().mean()
print(f'Dummy MAE (mean predictor) for Absorption max on test set: {mae:.2f} nm')
mae_emission = (test_dataset['Emission max (nm)'] - mean_emission).abs().mean()
print(f'Dummy MAE (mean predictor) for Emission max on test set: {mae_emission:.2f} nm')
```
